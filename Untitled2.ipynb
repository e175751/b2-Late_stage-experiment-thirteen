```python
import matplotlib.pyplot as plt 
import numpy as np
x=[]
y=[]
init =0.7
r=3.90
for i in range(0,101):
    if i%2==0: 
        x.append(init)
        init=r*(1-init)*init
        y.append(init) 
    else:
        x.append(init)
        y.append(init)
        
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel("Xn")
plt.ylabel("Xn+1")
x_fun = np.arange(0,1,0.001)
y_fun = 3.90 * (1-x_fun)*x_fun
x_line = np.arange(0, 1.1, 0.1)
y_line = x_line
plt.plot(x,y,x_line ,y_line ,x_fun ,y_fun)
plt.savefig("r_3.90.png")
```
